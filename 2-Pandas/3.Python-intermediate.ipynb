# List Comprehension
- shorter syntax when you will have to create a new list

```python
lst1 = ['Ganesh- Batch22', 'Abdul- Batch 22', 'Ajith- Batch22']
```

# - FILTER
- Function and condition- iterable
- Return an iterator yielding those items of iterable for which function(item) is true. If function is None, return the items that are true.

```python
def lower_marks(marks):
    return marks < 40
```

```python
lst = [1,2,3,4,5,68,8]
```

```python
list(filter(lower_marks, lst))
```

```python
list(filter(lambda x: x>40, lst))
```
