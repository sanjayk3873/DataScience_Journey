# LOC & ILOC

```python
movies_df = pd.read_csv("IMDB-Movie-Data.csv")
```

```python
movies_df.head(2)
```

```python
movies_df.tail()
```

```python
movies_df.loc[3:7,['Title','Genre','Director']]#here, the left to comma is row, and right to comma is column
```

```python
movies_df.loc[3:7:,['Title','Genre']] #here, the left to comma is row, and right to comma is column
```

```python
movies_df.loc[:,['Title','Genre']][3:7] #here, the left to comma is row, and right to comma is column
```

```python
movies_df.loc[3:7,3:6] #here, the left to comma is row, and right to comma is column
```

```python
movies_df.iloc[3:7,3:7] #here, the left to comma is row, and right to comma is column
```

```python
movies_df.iloc[0:7,[1,2,3]] #here, the left to comma is row, and right to comma is column
```

```python
movies_df.loc[:,'Actors'][0]
```

```python
movies_df.iloc[:,-2:-4] #here, the left to comma is row, and right to comma is column
```
